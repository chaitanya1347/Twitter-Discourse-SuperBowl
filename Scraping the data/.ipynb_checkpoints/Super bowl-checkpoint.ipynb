```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import os
```

```python
tweet_count = 1000000000
text_query = ["Super Bowl"]
since_date = "2021-01-01"
until_date = "2021-12-31"

# Using OS library to call CLI commands in Python
os.system('snscrape --jsonl --max-results {} --since {} twitter-search "{} until:{}"> text-query-tweet.json'.format(tweet_count, since_date, text_query, until_date))
```

```python
# Reads the json generated from the CLI command above and creates a pandas dataframe
for i in range (1,200):
    tweets_df = pd.read_json('text-query-tweet.json', lines=True)
```

```python
tweets_df
```

```python
tweets_df.drop(tweets_df.columns[11:],inplace =True,axis=1)
tweets_df.drop(["_type","user"],inplace =True,axis=1)
```

```python
tweets_df.to_csv("2021.csv")
```
